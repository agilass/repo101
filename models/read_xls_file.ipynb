```python
nb_rows=0
for ind in range(0,6):
    worksheet = workbook.sheet_by_index(ind)  #by index
    for i in range(15,worksheet.nrows):
        birth = dt.date.fromtimestamp((worksheet.cell(i, 5).value - 25569) * 86400.0).strftime('%Y/%m/%d')
        num=int(worksheet.cell(i, 26).value)
        uni = worksheet.cell(i, 23).value
        prenom =worksheet.cell(i, 12).value
        nom = worksheet.cell(i, 16).value
        genre = worksheet.cell(i, 11).value
        born_pl = worksheet.cell(i, 2).value
        classe = worksheet.cell(i, 30).value
        # test of connection
        cnx = mysql.connect(user='root', password = '', host ='127.0.0.1',database='mz2023')
        #Creating a cursor object using the cursor() method
        cursor = cnx.cursor()

        # Preparing SQL query to INSERT a record into the database.
        insert_stmt = (
           "INSERT INTO etudiants(num_classe , code , prenom , nom , sexe , dt_naissance , lieu_naissance ,classe)"
            " VALUES (%s, %s, %s, %s, %s, %s, %s, %s)"
        )
        data = (num, uni , prenom , nom , genre , birth  , born_pl , classe )
        try:
         # Executing the SQL command
            cursor.execute(insert_stmt, data)
        # Commit your changes in the database
            cnx.commit()
            nb_rows+=1
            cnx.close()
        except:
           # Rolling back in case of error
            cnx.rollback()
            print("Not completed successfully")
print(nb_rows)
        #Closing the connection
```
